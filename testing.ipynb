```python
import fastreg as fr
from fastreg import I, R, C, D, B
```

```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
# generate random testing data
data = fr.dataset(N=100_000, K1=10, K2=100, models=['linear', 'poisson'])
data
```

### Formulas

`Factor` â€” a column or expression

`Term` â€” a combination of `Factor`s

`Formula` â€” a combination of unique `Term`s

Built in `Factor`s are: `[R]eal`, `[C]ategorical`, `[I]ntercept`, `[D]emeaned`, and `[B]inned`

```python
# pure [R]eal (a factor)
R('x1')
```

```python
# interacted [R]eals (a term)
R('x1')*R('x2')
```

```python
# add in [I]ntercept and [C]ategorical (a formula)
I + R('x1')*R('x2') + C('id1')
```

```python
# add in [B]inned and [D]emeaned (a formula)
I + B('x1', 10)*R('x1') + D('x2', 'id1')
```

```python
# use general python expressions in factors
I + R('abs(x1)') + C('x2 > 0')
```

### Interfaces

```python
# pythonic formula
y = 'y'
x = I + R('x1') + R('x2') + C('id1') + C('id2')
print(f'{y} ~ {x}')
fr.ols(y=y, x=x, data=data)
```

```python
# R style formula
form = 'y ~ 1 + x1 + x2 + C(id1) + C(id2)'
fr.ols(formula=form, data=data)
```

```python
# tuple/list style formula
fr.ols(y='y', x=['1', 'x1', 'x2', C('id1'), C('id2')], data=data)
```

### Extras

```python
# demeaned in action (x1 unconditional, x2 conditional on id1)
fr.ols(y='y', x=I+D('x1')+D('x2', 'id1'), data=data)
```

```python
# quantile binning (x2 on deciles, self interacted)
fr.ols(y='y', x=I+R('x1')+B('x2', 10)*R('x2'), data=data)
```

```python
# hdfe â€” fast for one categorical term
fr.ols(y='y', x=I+R('x1')+R('x2'), hdfe=C('id1')*C('id2'), data=data)
```

```python
# absorbtion â€” fast but no std errors
fr.ols(y='y', x=I+R('x1')+R('x2'), absorb=C('id1')*C('id2'), data=data)
```

```python
# clustered standard errors (any term allowed)
fr.ols(y='y', x=I+R('x1')+R('x2')+C('id1')+C('id2'), cluster=C('id1')*C('id2'), data=data)
```

### Custom Factors

```python
import fastreg.formula as frm
```

```python
# simple mapper
@frm.factor()
def demean(x, k=2):
    return x - k*np.mean(x)
f = demean('x1', k=2)
print(f)
f.eval(data)
```

```python
# complex factor
@frm.factor(eval_args=(0, 1))
def cdemean(x, i=None, k=2):
    if i is None:
        means = np.mean(x)
    else:
        datf = pd.DataFrame({'vals': x, 'cond': i})
        cmean = datf.groupby('cond')['vals'].mean().rename('mean')
        datf = datf.join(cmean, on='cond')
        means = datf['mean'].values
    return x - k*means
cf = cdemean('x1', 'id1', k=2)
print(cf)
cf.eval(data)
```

```python
# named mapper
@frm.factor(name=lambda x, k: f'{x}-Î¼{k}')
def ndemean(x, k=2):
    return x - k*np.mean(x)
nf = ndemean('x1', k=2)
print(nf)
nf.eval(data)
```
